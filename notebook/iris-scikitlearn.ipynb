```python
from sklearn import datasets
```

```python
#Data from Iris of Scikit-learn
datasets.load_iris().data
```

```python
datasets.load_iris().target
```

```python
#Select just two first data characteristics (Sepal length and width)
iris = datasets.load_iris()
data_characteristics = iris.data[:, :2]
data_characteristics
```

```python
import matplotlib.pyplot as plt

#Scatter plot of the data
plt.figure(figsize=(10, 8))
plt.scatter(data_characteristics[:, 0], data_characteristics[:, 1], c=iris.target)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

#Decision Tree
plt.figure(figsize=(15,10))
dtc = DecisionTreeClassifier().fit(iris.data, iris.target)
plot_tree(dtc, filled=True)
```
