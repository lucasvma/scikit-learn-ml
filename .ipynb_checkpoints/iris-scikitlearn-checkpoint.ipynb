```python
from sklearn import datasets
```

```python
#Data from Iris of Scikit-learn
datasets.load_iris().data
```

```python
datasets.load_iris().target
```

```python
iris = datasets.load_iris()
#Select just two first data characteristics ()
data_C = iris.data[:, :2]
data_C
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.scatter(data_C[:, 0], data_C[:, 1], c=iris.target)
plt.xlabel('Comprimento da SÃ©pala')
plt.ylabel('Largura da SÃ©pala')
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

plt.figure(figsize=(15,10))
dtc = DecisionTreeClassifier().fit(iris.data, iris.target)
plot_tree(dtc, filled=True)
```
